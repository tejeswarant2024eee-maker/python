```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # Corrected import from kmeans to KMeans
x=[4,5,10,4,3,11,14,6,10,12]
y=[21,19,24,17,16,25,24,22,21,21]
data=list(zip(x,y))
# Corrected KMEANS to KMeans, N_CLUSTERS to n_clusters, and defined z with a value
kmeans = KMeans(n_clusters=3, random_state=0, n_init='auto') # Added random_state and n_init for reproducibility and to avoid future warnings
kmeans.fit(data)
plt.scatter(x,y,c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids') # Plotting centroids
plt.title('K-Means Clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
